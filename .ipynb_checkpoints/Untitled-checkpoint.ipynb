```python
import os
import glob

# Define paths
val_folder = "D:/Deep_Learning/Dataset/MRCNN/val/images"
train_folder = "D:/Deep_Learning/Dataset/MRCNN/train/images"

# Step 1: Extract base patterns from val folder (e.g., C10, D1_aug2, Nodefect_152)
val_patterns = set()

for val_file in os.listdir(val_folder):
    # Get the base name without extension (e.g., "C10" from "C10.png")
    base_name = os.path.splitext(val_file)[0]
    val_patterns.add(base_name)

print(f"Val patterns to match: {val_patterns}")

# Step 2: Search and delete matching files in train folder
for pattern in val_patterns:
    # Case 1: Files like "1032163_dataset..._C10.png" (long prefix + pattern)
    long_prefix_pattern = os.path.join(train_folder, f"*_{pattern}.*")
    matches = glob.glob(long_prefix_pattern)
    
    # Case 2: Files like "C10.png" (no prefix, exact match)
    exact_pattern = os.path.join(train_folder, f"{pattern}.*")
    matches += glob.glob(exact_pattern)
    
    # Delete all matches
    for match in matches:
        try:
            os.remove(match)
            print(f"🗑️ Deleted: {match}")
        except Exception as e:
            print(f"❌ Failed to delete {match}: {e}")

print("\nDeletion complete!")
```

```python
import os

# Define folders
image_folder = "D:/Deep_Learning/Dataset/MRCNN/train/images"
label_folder = "D:/Deep_Learning/Dataset/MRCNN/train/xml_labels"

# Get all image base names (without extension)
valid_image_basenames = set()
for file in os.listdir(image_folder):
    if file.lower().endswith((".jpg", ".jpeg", ".png")):
        basename = os.path.splitext(file)[0]
        valid_image_basenames.add(basename)

# Check label files
for label_file in os.listdir(label_folder):
    if not label_file.endswith(".xml"):
        continue

    label_basename = os.path.splitext(label_file)[0]
    if label_basename not in valid_image_basenames:
        full_path = os.path.join(label_folder, label_file)
        try:
            os.remove(full_path)
            print(f"🗑️ Deleted label (no matching image): {label_file}")
        except Exception as e:
            print(f"❌ Failed to delete {label_file}: {e}")
```
